```
df=pd.read_csv('/content/Restaurant_Reviews.tsv',sep='\t')
```

```
import nltk
import re
```

```
nltk.download('stopwords')
```

```
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```
corpus=[]
for i in range(1000):
  # Removing punctuations and other special characters and replace them with white space
  review=re.sub(pattern='[^a-zA-Z]',repl=' ',string=df['Review'][i])

  # Converting review to Lower case
  review=review.lower()

  # Tokenizing the review
  review_words=review.split()

  # Removing special characters
  review_words=[word for word in review_words if not word in set(stopwords.words('english'))]

  # Stemming of words
  review=[PorterStemmer().stem(word) for word in review_words]

  # Joining stemmed words
  review=' '.join(review)

  corpus.append(review)
```

```
corpus[0:5]
```

```
# Bag of Words Implementation
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500)
x=cv.fit_transform(corpus).toarray()
y=df.iloc[:,1].values
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```
from sklearn.naive_bayes import MultinomialNB
classifier=MultinomialNB()
classifier.fit(x_train,y_train)
```

```
pred=classifier.predict(x_test)
```

```
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,pred)
print(f'Accuracy is: {score}')
```

```
from sklearn.metrics import confusion_matrix
matrix=confusion_matrix(y_test,pred)
```

```
def sentiment(sample):
  sample=re.sub(pattern='[^a-zA-Z]',repl=' ',string=sample)
  sample=sample.lower()
  sample_words=sample.split()
  sample_words=[word for word in sample_words if not word in set(stopwords.words('english'))]
  result=[PorterStemmer().stem(word) for word in sample_words ]
  result=' '.join(result)

  check=cv.transform([result]).toarray()
  return classifier.predict(check)
```

```
# Predicting values
sample_review = 'Food was not tasty and the service was very slow.'

if sentiment(sample_review):
  print('This is a POSITIVE review.')
else:
  print('This is a NEGATIVE review!')
```
